```python
import numpy as np 
from numpy import random

# there's an example at the end of file
def CheckInput(x, y, f):
    if(len(x) != len(y)):
        print("Dữ liệu không thoả mã")
        return False
    if(len(f) < 1):
        print("Hệ hàm không thoả mãn")
        return False 
    return True

def least_square(x, y, f):
    if(not CheckInput(x,y,f)):
        print("Không thoả mãn điều kiện")
        return False;
    n = len(x);	
    m = len(f); fx = []
    x = np.array(x)
    for i in range(0, m):
        fx.append(f[i](x))
    one = np.array([1]*n).reshape(n,1)
    X = np.array(fx).T
    Y = np.array(y).reshape(n,1)
    XT = X.T
    XTX = XT.dot(X)
    XTY = XT.dot(Y)
    B = np.linalg.inv(XTX).dot(XTY)
    def new_f(x):
        s = 0
        for i in range(0, m):
            s += f[i](x)*B[i]
        return s
    return [new_f, B]



### An example
# x = np.arange(9) # x = [0,1,2,3,...8]
# y = np.sin(x)*4 + np.random.rand(9) + x**2 + 3

# """
# 	f1 = 1
# 	f2 = x
# 	f3 = sin(x)
# """
# f = [np.ones_like, lambda x: x, np.sin]

# [f, Bi] = least_square(x, y, f)
# print(Bi)
# b3 + x*b2 + b1*x^2 + b0*x^3
# poly_ls_get_formula(Bi,4) # print the formular out



# visualize
# import matplotlib.pyplot as plt 
# xx = np.linspace(min(x), max(x))
# plt.plot(x, y, 'ro')
# plt.plot(xx, f(xx), '--')
# plt.show()
```

```python
#Ridge Regression
def least_square_ridge(x, y, f,lamda):
    if(not CheckInput(x,y,f)):
        print("Condition(s) didn't meet")
        return False;
    n = len(x);	
    m = len(f); fx = []
    x = np.array(x)
    for i in range(0, m):
        fx.append(f[i](x))
    one = np.array([1]*n).reshape(n,1)
    X = np.array(fx).T
    Y = np.array(y).reshape(n,1)
    XT = X.T
    XTX = XT.dot(X) + lamda*np.ones(XT.dot(X).shape)
    XTY = XT.dot(Y)
    print(XTX)
    B = np.linalg.inv(XTX).dot(XTY)
    def new_f(x):
        s = 0
        for i in range(0, m):
            s += f[i](x)*B[i]
        return s
    return [new_f, B]
```

```python
#Chuyển ví dụ 1 về phụ thuộc tuyến tính sẽ không tìm được ma trận nghịch đảo với hàm Least_square cũ
x = np.array([0.78,1.56,2.34,3.12,3.81])
y = np.array([2.5,1.2,1.12,2.25,4.28])

"""
	f1 = 1
	f2 = x
	f3 = x^2
    f4 = 2x^2
    Đây là hệ hàm phụ thuộc tuyến tính
"""

f = [np.ones_like, lambda x: x, lambda x:x**2, lambda x:2*x**2]
[f, Bi] = least_square(x, y, f)

#loss function
print("Sai số là")
loss = np.sqrt(np.mean((f(x)-y)**2))
print(loss)
print("Hệ số B")
print(Bi)

# visualize
xx = np.linspace(min(x), max(x))
plt.plot(x, y, 'ro')
plt.plot(xx, f(xx), '--')
plt.show()
```

```python
#Chuyển ví dụ 1 về phụ thuộc tuyến tính và dùng Ridge Regression
x = np.array([0.78,1.56,2.34,3.12,3.81])
y = np.array([2.5,1.2,1.12,2.25,4.28])

"""
	f1 = 1
	f2 = x
	f3 = x^2
    f4 = 2*x^2
"""
lamda = 0.001

f = [np.ones_like, lambda x: x, lambda x:x**2,lambda x:2*x**2]

[f, Bi] = least_square_ridge(x, y, f,lamda)
#loss function
print("Sai số là")
loss = np.sqrt(np.mean((f(x)-y)**2))
print(loss)
print("Hệ số B")
print(Bi)

# visualize
xx = np.linspace(min(x), max(x))
plt.plot(x, y, 'ro')
plt.plot(xx, f(xx), '--')
plt.show()
```
